```python
from aocd import get_data
```

```python
data = get_data(day=1, year=2023)
data
```

```python
lines = data.split("\n")
for line in lines:
    print(line)
```

```python
def calibration_value(text: str) -> int:
    digits = "1234567890"
    first_dig = None
    second_dig = None
    for char in text:
        if char in digits:
            first_dig = char
            break
    for char in text[::-1]:
        if char in digits:
            second_dig = char
            break
    return int(first_dig) * 10 + int(second_dig)
calibration_value(lines[0])
```

```python
calibration_values = [calibration_value(line) for line in lines]
calibration_values
```

```python
sum = 0
for value in calibration_values:
    sum += value
sum
```

```python
from aocd import submit
```

```python
submit(sum, part="a", day=1, year=2023)
```

```python
def new_calibration_value(text: str) -> int:
    digits = list("1234567890")
    word_digits = ["one", "two", "three", "four", "five", "six", "seven", "eight", "nine", "zero"]
    digits += word_digits
    
    first_dig = None
    first_dig_index = len(text)
    second_dig = None
    second_dig_index = len(text)

    for digit in digits:
        try:
            index = text.index(digit)
        except ValueError:
            continue
        if index < first_dig_index:
            first_dig = digit
            first_dig_index = index
    for digit in digits:
        try:
            index = text[::-1].index(digit)
        except ValueError:
            continue
        if index < second_dig_index:
            second_dig = digit
            second_dig_index = index

    if first_dig in word_digits:
        first_dig = digits[word_digits.index(first_dig)]
    if second_dig in word_digits:
        second_dig = digits[word_digits.index(second_dig)]
    
    return int(first_dig) * 10 + int(second_dig)
new_calibration_value(lines[0])
```

```python
new_calibration_values = [new_calibration_value(line) for line in lines]
new_calibration_values
```

```python
new_sum = 0
for value in new_calibration_values:
    new_sum += value
new_sum
```

```python
submit(new_sum, part="b", day=1, year=2023)
```
